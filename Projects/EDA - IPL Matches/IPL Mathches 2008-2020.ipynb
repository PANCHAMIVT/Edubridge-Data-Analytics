## Visualisation

```python
plt.figure(figsize=(25,10))
plt.bar(d['winner'],d['venue'],color='red')
plt.title('winner Vs venue')
plt.xlabel('winner',fontsize=15)
plt.ylabel('venue',fontsize=15)
plt.show()
```

```python
d.pivot_table('year',columns='city',aggfunc='min')
sns.pairplot(d)
plt.show()
```

## Number of Matches per Venue

```python
sns.countplot('venue', data=d)
plt.xticks(rotation='vertical')
plt.show()
```

## Matches Played by Each Team

```python
x = d['team1'].value_counts()
y = d['team2'].value_counts()
(x+y).plot(kind='barh')
plt.show()
```

```python
print(d['winner'].value_counts())
plt.figure(figsize=(15,4))
sns.countplot(x=d['winner'])
plt.xticks(rotation=90)
plt.show()
```

## Matches Won by Each Team

```python
sns.countplot('winner', data=d)
plt.xticks(rotation='vertical')
plt.show()
```

## top 5 Players

```python
d4=d['player_of_match'].value_counts().head()

sns.barplot(x=d4.index,y=d4.values,data=d)
plt.title("Top 5")
plt.xticks(rotation=90)
plt.xlabel("Match Count")
plt.ylabel("Player")
plt.show()
```

```python
plt.title('Winner',fontsize=20)
d['winner'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(12,12))
plt.show()
```

```python
plt.title('toss_winner',fontsize=20)
d['toss_winner'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(22,12))
plt.show()
```

```python
sns.kdeplot(d['result_margin'])
plt.show()
```

## wins based on the toss_decision

```python
sns.set_style('whitegrid')
ax=sns.countplot(x='winner',hue='toss_decision',data=d,palette='rainbow')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
sns.violinplot(x='year',y='result_margin',data=d)
plt.show()
```
