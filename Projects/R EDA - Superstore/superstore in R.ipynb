```r
d=read.csv('Superstore.csv')
```

```r
sales_A=ggplot(d,aes(x=State,y=Sales,fill=State))+geom_col()
sales_A
```

```r
sales_A=ggplot(d,aes(x=State,y=Sales,fill=State))+geom_col()+coord_flip()+theme(legend.position='None')
sales_A
```

```r
sales_A=ggplot(d,aes(x=State,y=Sales,fill=State))+geom_col()+coord_flip()+theme(legend.position='None')+ggtitle('State wise analysis')
sales_A
```

```r
sales_A=ggplot(d,aes(x=State,y=Sales,fill=State))+geom_col()+coord_flip()+theme(legend.position='None',axis.text.y=element_text(size=6))+ggtitle('State wise analysis')
sales_A
```

```r
sales_A=ggplot(d,aes(x=Region,y=Sales,fill=Region))+geom_col()+coord_flip()+theme(legend.position='None',axis.text.y=element_text(size=6))+ggtitle('Region wise analysis')
sales_A
```
