```python
import pandas as pd
```

```python
d=pd.DataFrame({'Product_Name':productname,'Price':product_price,'Rating':rates})
d.head(10)                                
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.bar(d['Product_Name'],d['Rating'])
plt.title('Rating Vs Product_Name',fontsize=15,color='red')
plt.xlabel('Product_Name',fontsize=15)
plt.ylabel('Rating',fontsize=15)
plt.grid(True)
plt.show()
```

```python
print(d['Rating'].value_counts())
plt.figure(figsize=(15,4))
sns.countplot(x=d['Rating'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.bar(d['Product_Name'],d['Rating'],color="red")
plt.title('Rating Vs Product_Name',fontsize=15,color='red')
plt.xlabel('Product_Name',fontsize=15)
plt.ylabel('Rating',fontsize=15)
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='Rating',data=d,palette='coolwarm')
plt.show()
```

```python
plt.title('Rating',fontsize=20)
d['Rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Price',data=d,palette='tab10')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot('Price','Rating',data=d,color='r',label='Price')
plt.legend()
plt.show()
```
